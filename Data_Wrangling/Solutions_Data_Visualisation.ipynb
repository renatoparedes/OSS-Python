```python
## This code loads the data form the census.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option("display.max.columns", None)
df = pd.read_csv(
    "https://catalogodatos.gub.uy/dataset/b62bc86e-c4c2-4cb8-af18-388124eea2c9/resource/dec0a55d-1d86-4100-88ec-fc8dad06a363/download/microdatoscensopsicologia.csv"
)

df
```

## Data Wrangling

- There are some values registered as empty strings. Replace them with NaN values.

```python
# Write your solution here
df.replace(" ", np.nan, inplace=True)
df
```

- Now visualise the number of missing values in the data set

```python
# Write your solution here
import missingno as msno

msno.bar(df)
```

- Create a new dataset out of the subset of the columns: `Hs_promedio_trab_semana`, `Num_hijos`, `Sexo` `Próximos_tres_años_posgrado` and `Por_qué_no_posgrado`.

```python
# Write your solution here
new_df = df[
    [
        "Hs_promedio_trab_semana",
        "Num_hijos",
        "Sexo",
        "Próximos_tres_años_posgrado",
        "Por_qué_no_posgrado",
    ]
].copy()

new_df
```

- Now drop the rows with missing values.

```python
# Write your solution here
new_df.dropna(inplace=True)
new_df
```

- Verify that the data types of each variable are as follows:

| Variable 	| Type 	|
|:---------:	|:---------------------:	|
| Hs_promedio_trab_semana |        float64       	|
| Num_hijos |       int64       	|
|     Sexo     	|       string       	|
|     Próximos_tres_años_posgrado     	|       string       	|
|     Por_qué_no_posgrado       	|       string       	|


```python
# Write your solution here
new_df["Hs_promedio_trab_semana"] = new_df["Hs_promedio_trab_semana"].astype(
    float
)
new_df["Sexo"] = new_df["Sexo"].astype(str)
new_df["Próximos_tres_años_posgrado"] = new_df[
    "Próximos_tres_años_posgrado"
].astype(str)
new_df["Por_qué_no_posgrado"] = new_df["Por_qué_no_posgrado"].astype(str)
new_df.dtypes
```

- The variable `Por_qué_no_posgrado` is encoded as follows:

| Code	    |            Description            |
|:------:	|:--------------------------------:	|
|  3.231  	| Considera que no son necesarios  	|
|  3.232  	|  No tuvo acceso a la información 	|
|  3.233  	|        Razones familiares        	|
|  3.234  	|         Razones laborales        	|
|  3.235  	|        Razones económicas        	|
|  3.236  	|      Dificultad con idiomas      	|
|  3.237  	|           Otras causas           	|

Replace the values of this variable with the corresponding strings.

```python
# Write your solution here
codes = {
    "3.231": "Considera que no son necesarios",
    "3.232": "No tuvo acceso a la información",
    "3.233": "Razones familiares",
    "3.234": "Razones laborales",
    "3.235": "Razones económicas",
    "3.236": "Dificultad con idiomas",
    "3.237": "Otras causas",
}

new_df.replace({"Por_qué_no_posgrado": codes}, inplace=True)
new_df
```

- Repeat this procedure for the variables `Sexo`: 

| Sexo 	| Description 	|
|:---------:	|:---------------------:	|
|     0     	|        Femenino       	|
|     1     	|       Masculino       	|
|     2     	|       Otr@s       	|

and `Próximos_tres_años_posgrado`:

| Próximos_tres_años_posgrado 	| Description 	|
|:---------:	|:---------------------:	|
|     3.241     	|        Sí       	|
|     3.242     	|       No       	|

```python
# Write your solution here

sex_code = {"0": "Femenino", "1": "Masculino", "2": "Otr@s"}
pg_plans_code = {"3.241": "Sí", "3.242": "No"}

new_df.replace(
    {"Sexo": sex_code, "Próximos_tres_años_posgrado": pg_plans_code},
    inplace=True,
)
new_df
```

## Report on quantitative variables

## Plot the following graphs:

- Histogram illustrating the average working hours per week of the respondents using the variable `Hs_promedio_trab_semana`.

```python
# Write your solution here
sns.histplot(x="Hs_promedio_trab_semana", data=new_df)
```

- Box plot illustrating the number of children of respondents based on the variable `Num_hijos`.



```python
# Write your solution here
sns.boxplot(y="Num_hijos", data=new_df)
```

If you have reached this point, now edit the graphs so that the X and Y axes have appropriate titles in English. 

```python
# Write your solution here
histogram = sns.histplot(x="Hs_promedio_trab_semana", data=new_df)
histogram.set_xlabel("Average working hours per week", size=12)
histogram.set_ylabel("# Respondents", size=12)
plt.show()
```

```python
# Write your solution here
box = sns.boxplot(y="Num_hijos", data=new_df)
box.set_xlabel("Respondents", size=12)
box.set_ylabel("Number of children", size=12)
plt.show()
```

## Report on qualitative variables

## Plot the following graphs:

```python
# Write your solution here
sns.countplot(y="Por_qué_no_posgrado", data=new_df)
```

If you have reached this point, edit the graphics so that the category labels reflect their description in English.

```python
# Write your solution here
count = sns.countplot(y="Por_qué_no_posgrado", data=new_df)
count.set_xlabel("# Respondents")
count.set_ylabel("Reasons for not following graduate studies", size=10)
count.set_yticklabels(
    ["Family", "Economy", "Other", "Work", "No information", "Not necessary"],
    rotation=0,
)
plt.show()
```
